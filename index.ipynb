```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Loading the data
df = pd.read_csv('ames_train.csv')
```

```python
# Investigate the Data
df.info()
```

```python
# Investigating Distributions using scatter_matrix
pd.plotting.scatter_matrix(df[['LotArea','GrLivArea','SalePrice','YearBuilt','YrSold']], figsize = (15,15))
```

```python
# Create a plot that shows the SalesPrice Distribution
sale_price = df['SalePrice']

sale_price.plot.hist(bins = 30, figsize=(15,15), edgecolor = 'black')
plt.axvline(sale_price.mean(), color = 'black')
plt.axvline(sale_price.median(), color = 'black', ls = 'dashed')

```

```python
# Create a plot that shows the LotArea Distribution
LotArea = df['LotArea']

LotArea.plot.hist(bins = 30, figsize=(15,15), edgecolor = 'black')
plt.axvline(LotArea.mean(), color = 'black')
plt.axvline(LotArea.median(), color = 'black', ls = 'dashed')
```

```python
# Create a plot that shows the Distribution of the overall house condition
condition = df['OverallCond']

condition.plot.hist(bins = 30, figsize=(15,15), edgecolor = 'black')
plt.axvline(condition.mean(), color = 'black')
plt.axvline(condition.median(), color = 'black', ls = 'dashed')
```

```python
# Create a Box Plot for SalePrice
df['SalePrice'].plot.box()
```

```python
# Perform an Exploration of home values by age
df['age'] = df['YrSold'] - df['YearBuilt']

age = df['age']


print(age.corr(df['SalePrice']))


fig = df.plot.scatter('age', 'SalePrice', figsize=(10,10))
fig.set_aspect('auto')
fig.set_title('House Sales Price by age of Home')



#The correlation coefficient here shows -.52 which isn't strong, 
#but it is negative. It means that as the house gets older it has a negative effect on the sale price.
#The closer to 1 or -1 the coefficient is the stronger the trend line. This is not a very strong trend, but it is a slight trend.
#This is reflected in the scatter plot as we see the newer homes have more sales price outliers. This sseems to be a strong trend
#about age 60 where the houses plateau on decreasign value. More investigation would be required to understand why the trend levels off after age 60,
```
